```python
# # ライブラリのインポート
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler

%matplotlib inline 
```

```python
# # trainデータ
train = pd.read_csv('train.csv')
# # testデータ
test = pd.read_csv('test.csv')
```

```python
pd.set_option('display.max_columns', 100)
pd.set_option('display.max_rows', 100)
```

```python
df_train = pd.DataFrame(train)
```

```python
# 相関係数, SalePriceとの, 絶対値, ソート, 降順
df_train.corr().loc[:, ['SalePrice']].abs().sort_values('SalePrice', ascending = False)
# OverallQual GrLivArea GarageCars GarageArea TotalBsmtSF
```

```python
# 欠損値の確認
X = df_train.loc[:, ['OverallQual', 'GrLivArea', 'GarageCars', 'GarageArea', 'TotalBsmtSF']]
y = df_train.loc[:, ['SalePrice']]
print(X)
print(y)
# 欠損値なし
print(X.isnull().any())
```

```python
# plt.scatter(df_train.loc[:, ['OverallQual']].values, y.values)
sns.set()
cols = ['SalePrice','OverallQual', 'GrLivArea', 'GarageCars', 'GarageArea', 'TotalBsmtSF']
sns.pairplot(train[cols], height=2.5)
plt.show()
```

```python
# OveralQualで単回帰分析 + ホールドアウト法 + 多項式回帰(2次)
# 基底変換
from sklearn.preprocessing import PolynomialFeatures
quad = PolynomialFeatures(degree=2)
overal_qual = df_train.loc[:, ['OverallQual']].values
overal_qual_quad = quad.fit_transform(overal_qual)
overal_qual
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(overal_qual_quad, y, test_size = 0.3, random_state = 0)
```

```python
model_quad = LinearRegression()
model_quad.fit(X_train, y_train)
```

```python
plt.scatter(overal_qual, y, color='lightgray', label='data')
x = np.arange(0, 15, 1)[:, np.newaxis]
x_quad = quad.fit_transform(x)
plt.plot(x, model_quad.predict(x_quad), color='green', label='quad')
plt.show()
```

```python
df_test = pd.DataFrame(test)
x_pred = df_test.loc[:, ['OverallQual']]
ss = StandardScaler()
# 標準化して
x_pred_std = ss.fit_transform(x_pred)
# 多項式化して
x_pred_std_quad = quad.fit_transform(x_pred_std)
# 予測
y_pred = model_quad.predict(x_pred_std_quad).reshape(-1)
print(y_pred)
```
